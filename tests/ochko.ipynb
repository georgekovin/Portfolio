```python
#     
#   
def simple_decorator(func):
 
    # ,     
    #  func.      ,
    #    func,   decorated_function.
    #     ,  
    # *args  *kwargs.
    def decorated_function(*args, **kwargs):
        #   
        print("Input:")
        print("Positional:", args)
        print("Named:", kwargs)
        #    *args/**kwargs
        #     func
        result = func(*args, **kwargs)
        #    
        print("Result:", result)
        #    , 
        #      !
        return result
    #    
    # decorated_function
    return decorated_function


def root(value, n=2):
    result = value ** (1/n)
    return result
#   root    simple_decorator
decorated_root = simple_decorator(root)
#  decorated_root    
print(type(decorated_root))
#  :
# <class 'function'>


print(decorated_root(625, 4))
#  :
# Input:
# Positional: (625, 4)
# Named: {}
# Result: 5.0
# 5.0
```
