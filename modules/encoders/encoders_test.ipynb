# **Testing the DATA ENCODER function**

```python
import pandas as pd
from encoders import data_encoder
```

Dataset

```python
clothing_list = [
    ['xxs', 'dress'],
    ['xxs', 'skirt'],
    ['xs', 'dress'],
    ['s', 'skirt'],
    ['m', 'dress'],
    ['l', 'shirt'],
    ['s', 'coat'],
    ['m', 'coat'],
    ['xxl', 'shirt'],
    ['l', 'dress']
]

clothing = pd.DataFrame(clothing_list, columns = ['size',  'type'])

clothing
```

---

**With concatenation:**

Ordinal encoding

```python
data_encoder(data=clothing, encoder='ordinal')
```

OneHot encoding

```python
data_encoder(data=clothing, encoder='onehot') 
```

Binary encoding

```python
data_encoder(data=clothing, encoder='binary')
```

Unknown encoder

```python
data_encoder(data=clothing, encoder='python')
```
